# Finding overall sales during each month

```python
df1['Dateofbill'] = pd.to_datetime(df['Dateofbill'])
df1['Year'] = df1['Dateofbill'].dt.year
df1['Month'] = df1['Dateofbill'].dt.month
```

```python
df1['Dateofbill']
```

```python
monthly_sales = df1.groupby(df1['Dateofbill'].dt.to_period('M')).agg({'Final_Sales': 'sum'})
monthly_sales
```

```python
yearly_sales = df1.groupby(df1['Year']).agg({'Final_Sales': 'sum'})
yearly_sales
```

# Finding how the sub categories effect the overall sales .

```python
subcategory_sales = df1.groupby('SubCat')['Final_Sales'].sum()
subcategory_contribution = (subcategory_sales / df['Final_Sales'].sum()) * 100
subcategory_contribution = subcategory_contribution.sort_values(ascending=False)
subcategory_contribution
```

```python
customer_data = df1.groupby('Patient_ID').agg({
    'Final_Sales': 'sum',            # Total sales per customer
    'Dateofbill': 'nunique',         # Number of unique purchases per customer
})

# Calculate additional metrics
customer_data['Average_Purchase_Value'] = customer_data['Final_Sales'] / customer_data['Dateofbill']
customer_data['Lifetime_Value'] = customer_data['Average_Purchase_Value'] * customer_data['Dateofbill']

# Segment customers based on metrics
customer_segments = pd.cut(customer_data['Lifetime_Value'], bins=[0, 100, 500, float('inf')], labels=['Low-Value', 'Medium-Value', 'High-Value'])

# Add customer segments to the data
customer_data['Segment'] = customer_segments

# Print the customer data with segments
print(customer_data.head())
```
